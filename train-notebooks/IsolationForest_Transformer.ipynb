```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    mean_squared_error, mean_absolute_error,
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix
)
from sklearn.ensemble import IsolationForest
import tensorflow as tf
from tensorflow.keras.models import Model, Sequential, load_model, save_model
from tensorflow.keras.layers import Input, Dense, Dropout, LayerNormalization, MultiHeadAttention, GlobalAveragePooling1D, Embedding, Add
from tensorflow.keras.callbacks import EarlyStopping
import math
import pickle
import os
from datetime import datetime

# Create directories to store models and results
model_dir = 'models'
results_dir = 'results'
os.makedirs(model_dir, exist_ok=True)
os.makedirs(results_dir, exist_ok=True)

# Define headers for the dataset
header = [
    "timestamp",
    "cpu_usage",
    "top_1_cpu_proc_name", "top_1_cpu_proc_usage",
    "top_2_cpu_proc_name", "top_2_cpu_proc_usage",
    "top_3_cpu_proc_name", "top_3_cpu_proc_usage",
    "top_4_cpu_proc_name", "top_4_cpu_proc_usage",
    "top_5_cpu_proc_name", "top_5_cpu_proc_usage",
    "mem_usage",
    "top_1_mem_proc_name", "top_1_mem_proc_usage",
    "top_2_mem_proc_name", "top_2_mem_proc_usage",
    "top_3_mem_proc_name", "top_3_mem_proc_usage",
    "top_4_mem_proc_name", "top_4_mem_proc_usage",
    "top_5_mem_proc_name", "top_5_mem_proc_usage",
    "nginx_active_connections",
    "nginx_rps"
]

def load_and_preprocess_data(filepath, seq_length=10):
    # Load the dataset with semicolon separator and defined headers
    df = pd.read_csv(filepath, sep=';', names=header, header=None)

    # Convert timestamp to datetime for reference only (not used as features)
    df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')

    print(f"Dataset shape: {df.shape}")
    print(f"First 5 rows of the dataset:")
    display(df.head())

    numerical_columns = [
        'cpu_usage', 'top_1_cpu_proc_usage', 'top_2_cpu_proc_usage',
        'top_3_cpu_proc_usage', 'top_4_cpu_proc_usage', 'top_5_cpu_proc_usage',
        'mem_usage', 'top_1_mem_proc_usage', 'top_2_mem_proc_usage',
        'top_3_mem_proc_usage', 'top_4_mem_proc_usage', 'top_5_mem_proc_usage',
        'nginx_active_connections', 'nginx_rps'
    ]
    numerical_df = df[numerical_columns]

    print("\nMissing values in numerical columns:")
    display(numerical_df.isnull().sum())

    if numerical_df.isnull().any().any():
        print("Filling missing values with column means...")
        numerical_df = numerical_df.fillna(numerical_df.mean())

    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(numerical_df)
    scaled_df = pd.DataFrame(scaled_data, columns=numerical_df.columns)

    train_data, test_data = train_test_split(scaled_df, test_size=0.2, shuffle=False)

    print(f"Training data shape: {train_data.shape}")
    print(f"Testing data shape: {test_data.shape}")

    return {
        'df': df,
        'numerical_df': numerical_df,
        'scaled_df': scaled_df,
        'train_data': train_data,
        'test_data': test_data,
        'scaler': scaler,
        'numerical_columns': numerical_columns
    }

def visualize_data_distributions(numerical_df, columns=None):
    if columns is None:
        columns = numerical_df.columns

    n_features = len(columns)
    n_cols = 4
    n_rows = (n_features + n_cols - 1) // n_cols

    plt.figure(figsize=(20, 5 * n_rows))
    for i, feature in enumerate(columns):
        plt.subplot(n_rows, n_cols, i+1)
        sns.histplot(numerical_df[feature], kde=True)
        plt.title(f'Distribution of {feature}')

    plt.tight_layout()
    plt.savefig(f'{results_dir}/feature_distributions.png')
    plt.show()

def visualize_time_series(df, numerical_columns):
    key_metrics = ['cpu_usage', 'mem_usage', 'nginx_active_connections', 'nginx_rps']

    plt.figure(figsize=(15, 12))
    plt.subplot(3, 1, 1)
    plt.plot(df['timestamp'], df['cpu_usage'])
    plt.title('CPU Usage Over Time')
    plt.ylabel('CPU Usage (%)')
    plt.grid(True)

    plt.subplot(3, 1, 2)
    plt.plot(df['timestamp'], df['mem_usage'])
    plt.title('Memory Usage Over Time')
    plt.ylabel('Memory Usage (%)')
    plt.grid(True)

    plt.subplot(3, 1, 3)
    plt.plot(df['timestamp'], df['nginx_rps'], label='Requests Per Second')
    plt.plot(df['timestamp'], df['nginx_active_connections'], label='Active Connections')
    plt.title('Nginx Metrics Over Time')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig(f'{results_dir}/time_series.png')
    plt.show()

def plot_correlation_matrix(numerical_df):
    plt.figure(figsize=(14, 12))
    corr = numerical_df.corr()
    mask = np.triu(np.ones_like(corr, dtype=bool))
    sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm',
                vmin=-1, vmax=1, fmt='.2f', linewidths=0.5)
    plt.title('Feature Correlation Matrix')
    plt.tight_layout()
    plt.savefig(f'{results_dir}/correlation_matrix.png')
    plt.show()

# === ISOLATION FOREST ANOMALY DETECTION ===

def train_and_evaluate_isolation_forest(train_data, test_data):
    """
    Train and evaluate Isolation Forest for anomaly detection.
    Returns dict with model, threshold (outlier score), anomaly mask on test set.
    """
    iso_forest = IsolationForest(
        n_estimators=100,
        contamination=0.05,
        random_state=42,
        n_jobs=-1,
        verbose=1
    )
    iso_forest.fit(train_data)

    # Decision function returns anomaly scores (the lower, the more abnormal)
    test_scores = iso_forest.decision_function(test_data)
    
    # By default threshold is at 0.0 for sklearn's iso forest:
    # scores < 0 are anomalies.
    anomalies = test_scores < 0

    print(f"Number of anomalies detected: {np.sum(anomalies)}")
    print(f"Percentage of anomalies: {np.sum(anomalies) / len(test_scores) * 100:.2f}%")

    # Visualize anomaly score distribution and threshold
    plt.figure(figsize=(12, 6))
    plt.hist(test_scores, bins=50, alpha=0.75)
    
    threshold = 0.0
    plt.axvline(x=threshold, color='r', linestyle='--', label=f'Threshold: {threshold:.4f}')
    
    plt.title('Distribution of Anomaly Scores (Isolation Forest)')
    plt.xlabel('Anomaly Score')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.savefig(f'{results_dir}/isoforest_anomaly_score_distribution.png')
    plt.show()

    # Save model with pickle (not h5)
    with open(f'{model_dir}/isoforest_model.pkl', 'wb') as f:
        pickle.dump(iso_forest, f)
    
    np.save(f'{model_dir}/isoforest_threshold.npy', threshold)

    return {
        'model': iso_forest,
        'threshold': threshold,
        'test_scores': test_scores,
        'anomalies': anomalies
     }

# ==== TRANSFORMER FOR TIME SERIES ANOMALY DETECTION ====

def create_sequences(data, seq_length, labels=None):
    xs = []
    ys = []
    
    for i in range(len(data) - seq_length):
        x = data.iloc[i:(i + seq_length)].values
        xs.append(x)
        if labels is not None:
            y = labels[i + seq_length]
            ys.append(y)
    
    if labels is not None:
        return np.array(xs), np.array(ys)
    else:
        return np.array(xs)

def build_transformer_model(seq_length, n_features, d_model=64):
    """Minimal transformer encoder for sequence classification."""

    inp = Input(shape=(seq_length, n_features))                # (batch, seq_length, n_features)
    x = Dense(d_model)(inp)                                    # (batch, seq_length, d_model)

    # Learnable positional embedding
    pos_emb = Embedding(input_dim=seq_length, output_dim=d_model)
    positions = tf.range(start=0, limit=seq_length, delta=1)
    positions = pos_emb(positions)                             # (seq_length, d_model)
    x = x + positions                                          # Broadcasting adds to each batch

    # Transformer encoder block
    attn_output = MultiHeadAttention(num_heads=2, key_dim=d_model)(x,x)
    attn_output = Dropout(0.1)(attn_output)
    x2 = LayerNormalization(epsilon=1e-6)(x + attn_output)

    ff_output = Dense(64, activation="relu")(x2)
    ff_output = Dropout(0.1)(ff_output)
    x3 = LayerNormalization(epsilon=1e-6)(x2 + ff_output)

    pooled = GlobalAveragePooling1D()(x3)
    outp = Dense(16, activation='relu')(pooled)
    outp = Dense(1, activation='sigmoid')(outp)

    model = Model(inputs=inp, outputs=outp)
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
     )
    return model

def train_and_evaluate_transformer(test_data, anomalies, seq_length=10):
     y_pseudo = anomalies.astype(int)

     X_seq, y_seq = create_sequences(
         test_data.reset_index(drop=True),
         seq_length,
         y_pseudo
     )

     print(f"Transformer input shape: {X_seq.shape}")
     print(f"Transformer output shape: {y_seq.shape}")

     X_train, X_val, y_train, y_val = train_test_split(
         X_seq, y_seq, test_size=0.2, random_state=42
     )

     transformer_model = build_transformer_model(seq_length, X_train.shape[2])
     transformer_model.summary()

     early_stopping = EarlyStopping(
         monitor='val_loss',
         patience=10,
         restore_best_weights=True
     )

     history = transformer_model.fit(
         X_train, y_train,
         epochs=100,
         batch_size=32,
         validation_data=(X_val, y_val),
         callbacks=[early_stopping],
         verbose=1
     )

     # Plot training history (accuracy & loss)
     plt.figure(figsize=(15, 6))
     plt.subplot(1, 2, 1)
     plt.plot(history.history['accuracy'], label='Training Accuracy')
     plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
     plt.title('Transformer Model Accuracy')
     plt.xlabel('Epochs')
     plt.ylabel('Accuracy')
     plt.legend()
     plt.grid(True)

     plt.subplot(1, 2, 2)
     plt.plot(history.history['loss'], label='Training Loss')
     plt.plot(history.history['val_loss'], label='Validation Loss')
     plt.title('Transformer Model Loss')
     plt.xlabel('Epochs')
     plt.ylabel('Loss')
     plt.legend()
     plt.grid(True)

     plt.tight_layout()
     plt.savefig(f'{results_dir}/transformer_training.png')
     plt.show()

     # Evaluate Transformer model
     y_pred_proba = transformer_model.predict(X_val)
     y_pred = (y_pred_proba > 0.5).astype(int)

     accuracy = accuracy_score(y_val, y_pred)
     precision = precision_score(y_val, y_pred, zero_division=0)
     recall = recall_score(y_val, y_pred, zero_division=0)
     f1 = f1_score(y_val, y_pred, zero_division=0)

     print(f"Transformer Model Evaluation Metrics:")
     print(f"Accuracy: {accuracy:.4f}")
     print(f"Precision: {precision:.4f}")
     print(f"Recall: {recall:.4f}")
     print(f"F1 Score: {f1:.4f}")

     cm = confusion_matrix(y_val, y_pred)
     plt.figure(figsize=(8, 6))
     sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
     plt.title('Confusion Matrix')
     plt.xlabel('Predicted')
     plt.ylabel('Actual')
     plt.savefig(f'{results_dir}/transformer_confusion_matrix.png')
     plt.show()

     from sklearn.metrics import roc_curve, auc
     fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)
     roc_auc = auc(fpr, tpr)

     plt.figure(figsize=(8, 6))
     plt.plot(fpr, tpr, color='darkorange', lw=2,
              label=f'ROC curve (area = {roc_auc:.2f})')
     plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
     plt.xlim([0.0, 1.0])
     plt.ylim([0.0, 1.05])
     plt.xlabel('False Positive Rate')
     plt.ylabel('True Positive Rate')
     plt.title('Receiver Operating Characteristic (ROC) Curve')
     plt.legend(loc="lower right")
     plt.grid(True)
     plt.savefig(f'{results_dir}/transformer_roc_curve.png')
     plt.show()

     transformer_model.save(f'{model_dir}/transformer_model.h5')

     with open(f'{model_dir}/seq_length.txt', 'w') as f:
         f.write(str(seq_length))

     return {
         'model': transformer_model,
         'accuracy': accuracy,
         'precision': precision,
         'recall': recall,
         'f1': f1,
         'confusion_matrix': cm,
         'roc_auc': roc_auc,
         'history': history
      }

def process_csv_row(csv_row):
   values = csv_row.strip().split(';')
   data = dict(zip(header, values))

   numeric_fields = [
       "cpu_usage", "top_1_cpu_proc_usage", "top_2_cpu_proc_usage",
       "top_3_cpu_proc_usage", "top_4_cpu_proc_usage", "top_5_cpu_proc_usage",
       "mem_usage", "top_1_mem_proc_usage", "top_2_mem_proc_usage",
       "top_3_mem_proc_usage", "top_4_mem_proc_usage", "top_5_mem_proc_usage",
       "nginx_active_connections", "nginx_rps"
   ]
   for field in numeric_fields:
       try:
           data[field] = float(data[field])
       except (ValueError, KeyError):
           data[field] = 0.0

   return data

def detect_anomalies_realtime(data,
                             iso_forest=None,
                             threshold=None,
                             transformer_model=None,
                             sequence_buffer=None,
                             seq_length=None,
                             scaler=None):

   numerical_features = [
       'cpu_usage', 'top_1_cpu_proc_usage', 'top_2_cpu_proc_usage',
       'top_3_cpu_proc_usage', 'top_4_cpu_proc_usage', 'top_5_cpu_proc_usage',
       'mem_usage', 'top_1_mem_proc_usage', 'top_2_mem_proc_usage',
       'top_3_mem_proc_usage', 'top_4_mem_proc_usage', 'top_5_mem_proc_usage',
       'nginx_active_connections', 'nginx_rps'
   ]
   feature_values = [data.get(feature, 0.0) for feature in numerical_features]
   new_data_array = np.array([feature_values])

   if scaler is not None:
       new_data_scaled = scaler.transform(new_data_array)
   else:
       new_data_scaled = new_data_array

   # Isolation Forest anomaly score (lower < threshold means anomaly)
   anomaly_score = iso_forest.decision_function(new_data_scaled)[0]
   is_anomaly_iso_forest = anomaly_score < threshold

   result = {
       'timestamp': data.get('timestamp', datetime.now().isoformat()),
       'is_anomaly': bool(is_anomaly_iso_forest),
       'anomaly_score': float(anomaly_score),
       'threshold': float(threshold),
       'detection_method': 'isolation_forest'
   }

   # Transformer real-time prediction if buffer is available and enough points collected
   if (transformer_model is not None and sequence_buffer is not None and seq_length is not None and len(sequence_buffer) >= seq_length):
       sequence_buffer.append(new_data_scaled[0])
       if len(sequence_buffer) > seq_length:
           sequence_buffer.pop(0)

       input_seq = np.array([sequence_buffer[-seq_length:]])
       transformer_pred_proba = transformer_model.predict(input_seq)[0][0]
       is_anomaly_transformer = transformer_pred_proba > 0.5

       result['is_anomaly_transformer'] = bool(is_anomaly_transformer)
       result['transformer_confidence'] = float(transformer_pred_proba)
       result['detection_method'] = 'ensemble'
       result['is_anomaly'] = bool(is_anomaly_iso_forest or is_anomaly_transformer)

   return result

def detect_anomalies_from_csv(csv_row,
                             iso_forest=None,
                             threshold=None,
                             transformer_model=None,
                             sequence_buffer=None,
                             seq_length=None,
                             scaler=None):

   data = process_csv_row(csv_row)

   return detect_anomalies_realtime(
       data=data,
       iso_forest=iso_forest,
       threshold=threshold,
       transformer_model=transformer_model,
       sequence_buffer=sequence_buffer,
       seq_length=seq_length,
       scaler=scaler
   )

# Main execution flow
if __name__ == "__main__":
   print("Loading and preprocessing data...")
   data_dict = load_and_preprocess_data('new_system_stats.csv')

   with open(f'{model_dir}/scaler.pkl', 'wb') as f:
       pickle.dump(data_dict['scaler'], f)

   print("Visualizing data...")
   visualize_data_distributions(data_dict['numerical_df'])
   visualize_time_series(data_dict['df'], data_dict['numerical_columns'])
   plot_correlation_matrix(data_dict['numerical_df'])

   print("Training and evaluating Isolation Forest...")
   iso_results = train_and_evaluate_isolation_forest(
       data_dict['train_data'],
       data_dict['test_data']
   )

   print("Training and evaluating Transformer...")
   transformer_results = train_and_evaluate_transformer(
      data_dict['test_data'],
      iso_results['anomalies']
   )

   print("\nModel files saved to:", model_dir)
   print("Visualization results saved to:", results_dir)
```

```python
import numpy as np
import pickle
from tensorflow.keras.models import load_model

# Define the header (should match your training header)
header = [
    "timestamp",
    "cpu_usage",
    "top_1_cpu_proc_name", "top_1_cpu_proc_usage",
    "top_2_cpu_proc_name", "top_2_cpu_proc_usage",
    "top_3_cpu_proc_name", "top_3_cpu_proc_usage",
    "top_4_cpu_proc_name", "top_4_cpu_proc_usage",
    "top_5_cpu_proc_name", "top_5_cpu_proc_usage",
    "mem_usage",
    "top_1_mem_proc_name", "top_1_mem_proc_usage",
    "top_2_mem_proc_name", "top_2_mem_proc_usage",
    "top_3_mem_proc_name", "top_3_mem_proc_usage",
    "top_4_mem_proc_name", "top_4_mem_proc_usage",
    "top_5_mem_proc_name", "top_5_mem_proc_usage",
    "nginx_active_connections",
    "nginx_rps"
]

# --- Example usage ---
print("\nExample usage of the anomaly detection system:")

# Example data (as string)
timestamp = "2025-03-17T09:12:26"
cpu_usage = "50.50"
top_1_cpu_process_name = "python3 /etc/zabbix/templates/host_anomaly_detection/script.py"
top_1_cpu_process_usage = "38.90"
top_2_cpu_process_name = "sshd: root@notty "
top_2_cpu_process_usage = "6.50"
top_3_cpu_process_name = "/usr/lib/systemd/systemd --system --deserialize 17"
top_3_cpu_process_usage = "0.00"
top_4_cpu_process_name = "kthreadd"
top_4_cpu_process_usage = "0.00"
top_5_cpu_process_name = "kworker/0:0H"
top_5_cpu_process_usage = "0.00"
mem_usage = "52.60"
top_1_mem_process_name = "/usr/sbin/clamd -c /etc/clamd.d/amavisd.conf --foreground=yes"
top_1_mem_process_usage = "17.32"
top_2_mem_process_name = "/usr/sbin/mysqld --basedir=/usr"
top_2_mem_process_usage = "9.14"
top_3_mem_process_name = "/usr/sbin/amavisd (master)"
top_3_mem_process_usage = "1.58"
top_4_mem_process_name = "/usr/sbin/amavisd (virgin child)"
top_4_mem_process_usage = "1.57"
top_5_mem_process_name = "/usr/sbin/amavisd (virgin child)"
top_5_mem_process_usage = "1.57"
nginx_rps = "300"
nginx_active_connections = "0.67"

example_csv = f"{timestamp};{cpu_usage};{top_1_cpu_process_name};{top_1_cpu_process_usage};{top_2_cpu_process_name};{top_2_cpu_process_usage};{top_3_cpu_process_name};{top_3_cpu_process_usage};{top_4_cpu_process_name};{top_4_cpu_process_usage};{top_5_cpu_process_name};{top_5_cpu_process_usage};{mem_usage};{top_1_mem_process_name};{top_1_mem_process_usage};{top_2_mem_process_name};{top_2_mem_process_usage};{top_3_mem_process_name};{top_3_mem_process_usage};{top_4_mem_process_name};{top_4_mem_process_usage};{top_5_mem_process_name};{top_5_mem_process_usage};{nginx_rps};{nginx_active_connections}"

# --- Inference example ---

try:
    # Load scaler
    with open('models/scaler.pkl', 'rb') as f:
        scaler = pickle.load(f)

    # Load Isolation Forest
    with open('models/isoforest_model.pkl', 'rb') as f:
        iso_forest = pickle.load(f)

    threshold = np.load('models/isoforest_threshold.npy').item()

    # Load Transformer model
    from tensorflow.keras.models import load_model
    transformer_model = load_model('models/transformer_model.h5')

    seq_length = 10

    # Sequence buffer for streaming transformer inference (simulate real-time)
    sequence_buffer = []

    # --- Provide the function to process and detect anomalies ---
    def process_csv_row(csv_row):
        values = csv_row.strip().split(';')
        data = dict(zip(header, values))

        numeric_fields = [
            "cpu_usage", "top_1_cpu_proc_usage", "top_2_cpu_proc_usage",
            "top_3_cpu_proc_usage", "top_4_cpu_proc_usage", "top_5_cpu_proc_usage",
            "mem_usage", "top_1_mem_proc_usage", "top_2_mem_proc_usage",
            "top_3_mem_proc_usage", "top_4_mem_proc_usage", "top_5_mem_proc_usage",
            "nginx_active_connections", "nginx_rps"
        ]
        for field in numeric_fields:
            try:
                data[field] = float(data[field])
            except (ValueError, KeyError):
                data[field] = 0.0

        return data

    def detect_anomalies_realtime(data,
                                 iso_forest=None,
                                 threshold=None,
                                 transformer_model=None,
                                 sequence_buffer=None,
                                 seq_length=None,
                                 scaler=None):

        numerical_features = [
            'cpu_usage', 'top_1_cpu_proc_usage', 'top_2_cpu_proc_usage',
            'top_3_cpu_proc_usage', 'top_4_cpu_proc_usage', 'top_5_cpu_proc_usage',
            'mem_usage', 'top_1_mem_proc_usage', 'top_2_mem_proc_usage',
            'top_3_mem_proc_usage', 'top_4_mem_proc_usage', 'top_5_mem_proc_usage',
            'nginx_active_connections', 'nginx_rps'
        ]
        feature_values = [data.get(feature, 0.0) for feature in numerical_features]
        new_data_array = np.array([feature_values])
        if scaler is not None:
            new_data_scaled = scaler.transform(new_data_array)
        else:
            new_data_scaled = new_data_array

        # Isolation Forest anomaly score (lower < threshold means anomaly)
        anomaly_score = iso_forest.decision_function(new_data_scaled)[0]
        is_anomaly_iso_forest = anomaly_score < threshold

        result = {
            'timestamp': data.get('timestamp', datetime.now().isoformat()),
            'is_anomaly': bool(is_anomaly_iso_forest),
            'anomaly_score': float(anomaly_score),
            'threshold': float(threshold),
            'detection_method': 'isolation_forest'
        }

        # Transformer real-time prediction if buffer is available and enough points collected
        if (transformer_model is not None and sequence_buffer is not None and seq_length is not None):
            sequence_buffer.append(new_data_scaled[0])
            if len(sequence_buffer) > seq_length:
                sequence_buffer.pop(0)

            if len(sequence_buffer) == seq_length:
                input_seq = np.array([sequence_buffer])
                transformer_pred_proba = transformer_model.predict(input_seq, verbose=0)[0][0]
                is_anomaly_transformer = transformer_pred_proba > 0.5

                result['is_anomaly_transformer'] = bool(is_anomaly_transformer)
                result['transformer_confidence'] = float(transformer_pred_proba)
                result['detection_method'] = 'ensemble'
                result['is_anomaly'] = bool(is_anomaly_iso_forest or is_anomaly_transformer)

        return result

    def detect_anomalies_from_csv(csv_row,
                                 iso_forest=None,
                                 threshold=None,
                                 transformer_model=None,
                                 sequence_buffer=None,
                                 seq_length=None,
                                 scaler=None):
        data = process_csv_row(csv_row)
        return detect_anomalies_realtime(
            data=data,
            iso_forest=iso_forest,
            threshold=threshold,
            transformer_model=transformer_model,
            sequence_buffer=sequence_buffer,
            seq_length=seq_length,
            scaler=scaler
        )

    # --- Run detection on the example row ---
    result = detect_anomalies_from_csv(
        example_csv,
        iso_forest=iso_forest,
        threshold=threshold,
        transformer_model=transformer_model,
        sequence_buffer=sequence_buffer,
        seq_length=seq_length,
        scaler=scaler
    )

    # Print detection results
    print("\nAnomaly Detection Results:")
    print(f"Timestamp: {result['timestamp']}")
    print(f"Is anomaly: {result['is_anomaly']}")
    print(f"Detection method: {result['detection_method']}")
    print(f"Isolation Forest anomaly score: {result['anomaly_score']:.4f}")
    print(f"Threshold: {result['threshold']:.4f}")

    if 'is_anomaly_transformer' in result:
        print(f"Transformer anomaly detection: {result['is_anomaly_transformer']}")
        print(f"Transformer confidence: {result['transformer_confidence']:.4f}")

except Exception as e:
    print(f"Error in example usage: {e}")

print("\nSystem Deployment Guidelines:")
print("1. Load the trained Isolation Forest and Transformer models")
print("2. Initialize an empty sequence buffer for Transformer predictions")
print("3. For each new data point:")
print("   - Process the raw CSV data")
print("   - Extract and normalize features")
print("   - Detect anomalies using both models")
print("   - Log or alert if an anomaly is detected")
print("   - Update the sequence buffer for future predictions")

# Performance evaluation summary (example; adjust according to your variable names)
try:
    # Load test results for summary (assuming iso_results and transformer_results exist)
    print("\nPerformance Summary:")
    print(f"Isolation Forest threshold: {threshold:.4f}")
    print(f"Percentage of anomalies detected: {np.sum(iso_results['anomalies']) / len(iso_results['anomalies']) * 100:.2f}%")
    print(f"Transformer model final accuracy: {transformer_results['history'].history['val_accuracy'][-1]:.4f}")
    print(f"Transformer model final loss: {transformer_results['history'].history['val_loss'][-1]:.4f}")
    print(f"Transformer model F1 score: {transformer_results['f1']:.4f}")
except Exception as e:
    print(f"(Performance summary unavailable) Error: {e}")
```
